# Machine Learning when the machine doesn't want to learn 
___

Using the dataset from the first runs with 3800 runs.

Each individual numpy in <code>np_dir</code> is a $[30,20,20]$ array where the deposition image at time t is $[:,:,\text{t}]$. The name of each numpy contains the corresponding run parameters like

``` <RUN_ID>_<wind_velocity>_<wind_angle>_<nozzle_exit_velocity>_<nozzle_polar_angle>_<nozzle_azimuthal_angle>.npy ```

The total injected mass per simulation is 0.095 kg.

Importing by getting a list of all numpys in <code>np_dir</code>, then looping through the list - loading each numpy, extracting and normalising the final image and storing the run parameters. Then <code>np._stack()</code>-ing the input/ouput arrays, with one output flatten and one still in image form.



````python
import os
import numpy as np
from skimage import data
from skimage.filters import threshold_otsu

# general run param definition
injection_volume = 0.095
steadystate_time = 12

# get list of all numpys
%cd 'total'
np_sets = [".".join(f.split(".")[:-1]) for f in os.listdir() if os.path.isfile(f)]
thresh = injection_volume / ( 30*20 )
# iterate over all np_sets, import and normalise, flatten them, catenate into one array and extract the input params from the name:
for idx, dataframe in enumerate(np_sets):
    
    df = np.load( (dataframe + '.npy') )
    df = (df[:,:,steadystate_time] - df[:,:,steadystate_time-1]) / 0.02 
    
    thresh_ot = threshold_otsu(df)
    df_ot = df > thresh_ot
    df_f = df.flatten()
    df_th = df > thresh
    nms = dataframe.split('_')
    run_params = np.array([nms[1], nms[2], nms[3], nms[4], nms[5]], 'f')
    
    if idx == 0:
        inputs = run_params
        outputs_f = df.flatten()
        outputs   = df
        out_deposition = df.sum()
        out_th = df_th
        out_th_f = df_th.flatten()
        out_ot = df_ot
        out_ot_f = df_ot.flatten()
    else:
        inputs = np.vstack((run_params, inputs))
        outputs_f = np.vstack((outputs_f, df.flatten()))
        outputs = np.dstack((outputs, df))
        out_deposition = np.vstack((out_deposition, df.sum()))
        out_th = np.dstack((out_th, df_th))
        out_th_f = np.vstack((out_th_f, df_th.flatten()))
        out_ot = np.dstack((out_ot, df_ot))
        out_ot_f = np.vstack((out_ot_f, df_ot.flatten()))
        
%cd "../"

outputs = np.swapaxes(outputs, 0, 2)
outputs = np.swapaxes(outputs, 1, 2)

out_ot = np.swapaxes(out_ot, 0, 2)
out_ot = np.swapaxes(out_ot, 1, 2)

print('input shape:', np.shape(inputs))
print('output shape:', np.shape(outputs))
````

start by defining a few metrics and a plot function:

````python
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import random

def get_metrics(y_true, y_pred):
    r2 = r2_score(y_true, y_pred)
    mean_abs = mean_absolute_error(y_true, y_pred)
    mean_sq = mean_squared_error(y_true, y_pred)
    
    print('R2: %0.3f \n mean absolute error: %0.3f \n mean square error: %0.3f' % (r2, mean_abs, mean_sq))
    return
 
def plot_pred(CLF):
    views = random.choices(idx_test, k = 10)
    pred = CLF.predict(inputs[views, :])

    fig, ax = plt.subplots(2,10, figsize = (18,6))
    
    for idx, case in enumerate(views):
        ax[0, idx].imshow(outputs[case, :, :], clim = (0,0.005), cmap = 'hot')
        ax[1, idx].imshow(pred[idx, :].reshape(30,20), clim = (0,0.005), cmap = 'hot')
        ax[1, idx].get_xaxis().set_visible(False)
        ax[1, idx].get_yaxis().set_visible(False)
        ax[0, idx].get_xaxis().set_visible(False)
        ax[0, idx].get_yaxis().set_visible(False)

    return
````

````python
## get a conv
from keras.layers import Input, Dense, Conv2D, MaxPooling2D, UpSampling2D, AveragePooling2D
from keras.models import Model

input_img = Input(shape=(30, 20, 1))  


############
# Encoding #
############

# Conv1 #
x = Conv2D(filters = 16, kernel_size = (3, 3), activation='relu', padding='same')(input_img)
x = AveragePooling2D(pool_size = (3, 2), padding='same')(x)

# Conv2 #
x = Conv2D(filters = 8, kernel_size = (3, 3), activation='relu', padding='same')(x)
x = AveragePooling2D(pool_size = (2, 2), padding='same')(x) 

# Conv 3 #
x = Conv2D(filters = 8, kernel_size = (3, 3), activation='relu', padding='same')(x)
encoded = AveragePooling2D(pool_size = (2, 2), padding='same')(x)


############
# Decoding #
############

# DeConv1
x = Conv2D(8, (3, 3), activation='relu', padding='same')(encoded)
x = UpSampling2D((2, 2))(x)

# DeConv2
x = Conv2D(8, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)

# Deconv3
x = Conv2D(16, (3, 3), activation='relu')(x)
x = UpSampling2D((3, 2))(x)
decoded = Conv2D(1, (3, 3), activation='tanh', padding='same')(x)



# Train the model

autoencoder = Model(input_img, decoded)
autoencoder.compile(optimizer='adadelta', loss='MSE')

#autoencoder.fit(x_train, x_train)
````

````python
from keras.layers import Input, Dense, Conv2D, MaxPooling2D, UpSampling2D, AveragePooling2D
from keras.models import Model

input_img = Input(shape=(30, 20, 1))  


############
# Encoding #
############

# Conv1 #
x = Conv2D(filters = 2, kernel_size = (1, 1), activation='relu', padding='same')(input_img)
x = Conv2D(filters = 2, kernel_size = (2, 2), activation='relu', padding='same')(x)
x = AveragePooling2D(pool_size = (3, 2), padding='same')(x)

# Conv2 #
x = Conv2D(filters = 8, kernel_size = (2, 2), activation='relu', padding='same')(x)
x = Conv2D(filters = 8, kernel_size = (2, 2), activation='relu', padding='same')(x)
x = AveragePooling2D(pool_size = (2, 2), padding='same')(x) 

# Conv 3 #
x = Conv2D(filters = 32, kernel_size = (1, 1), activation='relu', padding='same')(x)
encoded = Conv2D(filters = 32, kernel_size = (1, 1), activation='relu', padding='same')(x)

############
# Decoding #
############

# DeConv1
x = Conv2D(32, (1, 1), activation='relu', padding='same')(encoded)
x = Conv2D(32, (1, 1), activation='relu', padding='same')(x)

# DeConv2
x = UpSampling2D((2, 2))(x)
x = Conv2D(filters = 8, kernel_size = (2, 2), activation='relu', padding='same')(x)
x = Conv2D(filters = 8, kernel_size = (2, 2), activation='relu', padding='same')(x)

# Deconv3
x = UpSampling2D((3, 2))(x)
x = Conv2D(filters = 2, kernel_size = (2, 2), activation='relu', padding='same')(x)
x = Conv2D(filters = 2, kernel_size = (2, 2), activation='relu', padding='same')(x)
decoded = Conv2D(filters = 1, kernel_size = (1, 1), activation='relu', padding='same')(x)


# Train the model

autoencoder = Model(input_img, decoded)
autoencoder.compile(optimizer='adadelta', loss='MSE')

autoencoder.fit(x_train, x_train,
                epochs=200,
                batch_size = 30,
                shuffle=True
               )
````
